```python
import helper_module as hm

card_base = hm.BASE_PATH.joinpath("packages/tb_core/lib/src/cards")
log_base = hm.BASE_PATH.joinpath("packages/tb_core/lib/src/models/game/logging")
```

```python
cards = hm.get_yaml_dict(
    hm.ASSET_PATH.joinpath("localizables/en/card_names.yaml")
).keys()



ftext = """



import '../models/game/game.dart';



import 'card.dart';



import 'card_catalog.dart';




/// The {card_name_caps} card.



class {card_name_caps} extends Card {{



  /// Creates a [{card_name_caps}].



  {card_name_caps}() : super(key: GameCard.{card_name});
}}

"""



for card_name in cards:

    card_name_caps = card_name.capitalize()

    text = ftext.format(card_name_caps=card_name_caps, card_name=card_name)

    fpath = card_base.joinpath(f"{card_name}.dart")

    if fpath.exists():

        print(f"File {fpath} already exists")

        continue

    with fpath.open("w") as f:

        f.write(text)

    print(f"Created file {fpath}")
len(cards)
```

```python
log_type_ftext = """
import 'package:json_annotation/json_annotation.dart';

import '../../../wrapper/tr_object.dart';
import '../game.dart';
import 'log_entry.dart';
import 'log_entry_type.dart';

part '{log_type_lower}.g.dart';

@JsonSerializable()

///
class {log_type_name_caps} extends LogEntry {{
  /// Creates a [{log_type_name_caps}].
  {log_type_name_caps}({{
    int? indentLevel,
  }}) : super(
          entryType: LogEntryType.{log_type_name},
          indentLevel: indentLevel ?? 0,
        );

  /// Creates a [{log_type_name_caps}] from a JSON map.
  factory {log_type_name_caps}.fromJson(Map<String, dynamic> json) =>
      _${log_type_name_caps}FromJson(json);

  @override
  Map<String, dynamic> toJson() => _${log_type_name_caps}ToJson(this)
    ..addEntries([
      MapEntry('entryType', entryType.name),
    ]);

  @override
  TrObject getDescription(Game game) => TrObject(localizedKey);
}}
"""

dot_g_ftext = """// GENERATED CODE - DO NOT MODIFY BY HAND

part of '{log_type_lower}.dart';

// **************************************************************************
// JsonSerializableGenerator
// **************************************************************************

LogUnitPlaced _${log_type_name_caps}FromJson(Map json) => Log{log_type_name_caps}(
      indentLevel: json['indentLevel'] as int?,
    );

Map<String, dynamic> _${log_type_name_caps}ToJson({log_type_name_caps} instance) =>
    <String, dynamic>{{
      'indentLevel': instance.indentLevel,
    }};
"""

new_log_types = ["unit_mod_single", "unit_mod_multi"]


def convert_snake_to_upper_camel(string: str) -> str:
    return "".join([s.capitalize() for s in string.split("_")])


for log_type in new_log_types:
    log_name = convert_snake_to_upper_camel(log_type)
    log_name_caps = "Log" + log_name
    log_name = log_name[0].lower() + log_name[1:]
    text = log_type_ftext.format(
        log_type_name_caps=log_name_caps,
        log_type_name=log_name,
        log_type_lower=log_type,
    )
    fpath = log_base.joinpath(f"{log_type}.dart")
    if fpath.exists():
        print(f"File {fpath} already exists")
        continue
    with fpath.open("w") as f:
        f.write(text)
    print(f"Created file for {log_type}")
    dot_g_text = dot_g_ftext.format(
        log_type_name_caps=log_name_caps,
        log_type_name=log_name,
        log_type_lower=log_type,
    )
    dot_g_fpath = log_base.joinpath(f"{log_type}.g.dart")
    if dot_g_fpath.exists():
        print(f"File {dot_g_fpath} already exists")
        continue
    with dot_g_fpath.open("w") as f:
        f.write(dot_g_text)
    print(f"Created generated file {log_type}")
```
